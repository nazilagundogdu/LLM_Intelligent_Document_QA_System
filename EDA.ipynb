## 3. Tokenizing and Word Counting the `summary` with `NLTK`

One area of interest is the "abstract" of the papers, stored in the `summmary` column/row of the dataframe. First, I isolate this information in a `pandas` `Series`, assigning it to the variable name "summary" below.

```python
summary = df.loc['summary',:]
```

Next, I prepare the data for tokenization. To do so, I download the stopwords from `NLTK`, as these words, although frequent, add little value to the semantics of the text, which is what I am most interested in. 

I also do not need punctuation marks. In order to get rid of them, I use another `NLTK` package, i.e. `RegexpTokenizer`, which allows me to use regular expressions while tokenizing the text. 

All these are imported in the first section of this notebook. 

```python
nltk.download('stopwords')

# get English stopwords 
stop_words = set(stopwords.words('english'))

# initiate a tokenizer that would only take into account "words" 
tokenizer = RegexpTokenizer(r'\w+')
```

The next step is to create a holder, here I chose a list, for all the "absract" data in the dataframe. 

```python
# define a list to store the tokens in 
all_tokens= []
for i in range(len(summary)):
    text = summary[i]
    
    # lower the text and tokenize it
    tokens = tokenizer.tokenize(text.lower())

    # remove stop words
    filtered_tokens = [word for word in tokens if word not in stop_words]
    all_tokens.append(filtered_tokens)
```

<a id='visualize'></a>

### 3.1. Visualize

```python
token_lengths = [len(item) for item in all_tokens]

plt.hist(token_lengths)

plt.title("Distribution of Token Lengths")
plt.xlabel("Token Length")
plt.ylabel("Frequency")

plt.axvline(np.mean(token_lengths), color = 'red', linestyle = '--', label = 'mean')
plt.axvline(np.mean(token_lengths) - 2 * np.std(token_lengths), color = 'green', linestyle = '--', label = '2std')
plt.axvline(np.mean(token_lengths) + 2 * np.std(token_lengths), color = 'green', linestyle = '--')

plt.legend()
plt.show();
```

The abstract length shows a normal distributionm, with the following statistics:

```python
max(token_lengths)
```

```python
np.mean(token_lengths)
```

```python
np.std(token_lengths)
```

<a id='word.freq'></a>
### 3.1.1. Word freuency

```python
all_tokens_flat = []

for sublist in all_tokens:
    for item in sublist:
        all_tokens_flat.append(item)
```

```python
fdist = FreqDist(all_tokens_flat)
fdist
```

The result is not surprising, with "ad", short for "Alzheimer Disease" being the most frequent word used in the papers about "ad"!

```python
most_common = dict(Counter(fdist).most_common(30))

plt.figure(figsize=(12, 6))
plt.bar(most_common.keys(), most_common.values())
plt.xticks(rotation=45)
plt.title("Top 30 Most Frequent Items")
plt.tight_layout()
plt.show()
```

```python

wordcloud = WordCloud(width=800, height=400, background_color='white').generate(' '.join(all_tokens_flat))

plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')  
plt.title("Word Cloud for Articles on Alzheimer Disease")
plt.show()
```

## 4. Number of authors



```python
authors = df.loc['authors',:]
```

```python
authors_count = []
for author in authors:
    authors_count.append(len(author))
```

```python
authors_count_unique = Counter(authors_count)
```

```python
plt.bar(authors_count_unique.keys(), authors_count_unique.values())

plt.xlabel('Number of Authors')
plt.ylabel('Frequency')
plt.title('Frequency of Author Count')

plt.show();
```

This plot shows a number close to 100 for the number of authors for one paper. It is not common to have that many authors for a paper. Let's check to see if we can verify that. First, start with some stats.

```python
max(authors_count)
```

```python
np.mean(authors_count)
```

```python
np.median(authors_count)
```

## 5. Number of publications per year

```python
df.head()
```

```python
publication = pd.DataFrame(pd.to_datetime(df.loc['published']))
```

```python
publication['year'] = publication['published'].dt.year
```

```python
counts = publication['year'].value_counts()
```

```python
plt.plot(publication['year'].sort_values().unique(), counts.sort_values())
plt.xlabel('Year')
plt.ylabel('Number of Publications')

plt.title('Number of Publications per Year')

plt.show()
```

The number of publications seems to be on the rise. It is hard to say for sure, because of the following 2 reasons:

1. There are other journals we have not looked at. It could be the case that `arXiv.org` was not the most common journal for this kind of work.
2. Due to computational resources limitations, I only extracted the data for 1000 random papers. It might be the case that the population the data was scraped from was concentrated in the more recent years.
